```python
import pandas as pd
```

```python
#df = pd.read_csv('./data/smokers_2000_2020.csv')
df_2020 = pd.read_csv('./data/data2020_429.csv', low_memory=False) # df of follow up study
df_2020 = df_2020[df_2020['soziodemographie_timestamp'].notna()] # drop columns of participants only from 2000
df_2000 = pd.read_csv('./data/data2000.csv', low_memory=False) # all participants of first study
df_smokers = pd.read_csv('./data/smokers_2000_2020.csv', low_memory=False)
df_smokers_selected_items = df_smokers # only selected items: AuDit etc.
df_smokers = df_smokers[['participant_id', 'smoker_steady', 'nonsmoker_steady', 'quit', 'start']]
df_smokers.head()
```

We merge the two dataframes from 2000 and 2020 by participant IDs of the 428 participants of the follow up study

```python
df = df_2020.merge(df_2000, how='left', on='participant_id') # merge data from 2000 and 2020
df = df.merge(df_smokers, how='left', on='participant_id') # merge 2020 frame with smoking_status
```

```python
df_smoker_steady = df[df["smoker_steady"] == 1]
df_smoker_steady = df_smoker_steady.drop(columns=['quit', 'start', 'smoker_steady', 'nonsmoker_steady'])
df_smoker_steady.head()
```

```python
df_nonsmoker_steady = df[df["nonsmoker_steady"] == 1]
df_nonsmoker_steady = df_nonsmoker_steady.drop(columns=['quit', 'start', 'smoker_steady', 'nonsmoker_steady'])
df_nonsmoker_steady.head()
```

```python
df_nonsmoker_steady.shape
```

```python
df_quit = df[df["quit"] == 1]
df_quit = df_quit.drop(columns=['quit', 'start', 'smoker_steady', 'nonsmoker_steady'])
df_quit.head()
```

```python
df.quit.shape
```

For the three groups: 'smoker_steady', 'nonsmoker_steady' and 'quit', we want to look at the correlations and do three heatmaps

```python
df_smoker_steady_cor = df_smoker_steady.corr()
df_smoker_steady_cor.to_csv("./data/smoker_steady_cor.csv")

df_nonsmoker_steady_cor = df_nonsmoker_steady.corr()
df_nonsmoker_steady_cor.to_csv("./data/nonsmoker_steady_cor.csv")
                            
df_quit_cor = df_quit.corr()
df_quit_cor.to_csv("./data/quit_cor.csv")
```

```python
df_ss_selected = df_smokers_selected_items[df_smokers_selected_items["smoker_steady"] == 1]
df_ss_selected = df_ss_selected.drop(columns=['quit', 'start', 'smoker_steady', 'nonsmoker_steady'])

df_ns_selected = df_smokers_selected_items[df_smokers_selected_items["nonsmoker_steady"] == 1]
df_ns_selected = df_ns_selected.drop(columns=['quit', 'start', 'smoker_steady', 'nonsmoker_steady'])

df_quit_selected = df_smokers_selected_items[df_smokers_selected_items["quit"] == 1]
df_quit_selected = df_quit_selected.drop(columns=['quit', 'start', 'smoker_steady', 'nonsmoker_steady'])
```

```python
df_ss_selected_cor = df_ss_selected.corr()
df_ns_selected_cor = df_ns_selected.corr()
df_quit_selected_cor = df_quit_selected.corr()
```

```python
import seaborn as sns

ss_heatmap = sns.heatmap(df_ss_selected_cor, xticklabels=df_ss_selected_cor.columns, yticklabels=df_ss_selected_cor.columns)
```

```python
from pylab import savefig

figure = ss_heatmap.get_figure() 
figure.savefig('ss_heatmap.png', dpi=1500)
```

```python
ns_heatmap = sns.heatmap(df_ns_selected_cor, xticklabels=df_ns_selected_cor.columns, yticklabels=df_ns_selected_cor.columns)
```

```python
figure = ns_heatmap.get_figure() 
figure.savefig('ns_heatmap.png', dpi=1500)
```

```python
quit_heatmap = sns.heatmap(df_quit_selected_cor, xticklabels=df_quit_selected_cor.columns, yticklabels=df_ns_selected_cor.columns)
```

```python
figure = quit_heatmap.get_figure() 
figure.savefig('quit_heatmap.png', dpi=1500)
```
